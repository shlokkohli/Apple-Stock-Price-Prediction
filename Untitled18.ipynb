```python
# Import packages
import kagglehub

# Download Apple stock dataset
path = kagglehub.dataset_download("varpit94/apple-stock-data-updated-till-22jun2021")
file_path = f"{path}/AAPL.csv"

# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")

np.random.seed(42)
tf.random.set_seed(42)

# Load stock data
def get_stock_data_from_csv(file_path):
    df = pd.read_csv(file_path)
    df.rename(columns={'Date': 'date', 'Close': 'price'}, inplace=True)
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date')
    df.set_index('date', inplace=True)
    return df[['price']]

# Create LSTM dataset
def create_dataset(dataset, time_step=1):
    X, Y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        Y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(Y)

# Build LSTM model
def build_lstm_model(X_train):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(50))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Evaluation metrics
def calculate_metrics(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return {'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'MAPE': mape}

# Main pipeline
def main():
    time_step = 15
    train_ratio = 0.8

    df = get_stock_data_from_csv(file_path)
    df.dropna(inplace=True)

    # Normalize for LSTM
    dataset = df.values
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)

    train_size = int(len(scaled_data) * train_ratio)
    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size:]

    X_train, y_train = create_dataset(train_data, time_step)
    X_test, y_test = create_dataset(test_data, time_step)
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    # Train LSTM
    print("Training LSTM...")
    lstm_model = build_lstm_model(X_train)
    lstm_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32, verbose=1)

    # LSTM predictions
    lstm_train_pred = lstm_model.predict(X_train)
    lstm_test_pred = lstm_model.predict(X_test)
    lstm_train_pred = scaler.inverse_transform(lstm_train_pred)
    lstm_test_pred = scaler.inverse_transform(lstm_test_pred)
    y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    lstm_metrics = calculate_metrics(y_test_inv, lstm_test_pred)
    print("\nLSTM Test Metrics:")
    for k, v in lstm_metrics.items():
        print(f"{k}: {v:.4f}")

    # ARIMA
    print("\nTraining ARIMA...")
    df_train = df.iloc[:train_size]
    df_test = df.iloc[train_size:]
    arima_order = (5, 1, 0)  # Manual order
    arima_model = ARIMA(df_train['price'], order=arima_order).fit()
    arima_forecast = arima_model.forecast(steps=len(df_test))
    arima_metrics = calculate_metrics(df_test['price'].values, arima_forecast)
    print("\nARIMA Test Metrics:")
    for k, v in arima_metrics.items():
        print(f"{k}: {v:.4f}")

    # SARIMA
    print("\nTraining SARIMA...")
    sarima_order = (1, 1, 1)
    seasonal_order = (1, 1, 1, 12)  # Adjust s if your data is not monthly
    sarima_model = SARIMAX(df_train['price'],
                           order=sarima_order,
                           seasonal_order=seasonal_order).fit()
    sarima_forecast = sarima_model.forecast(steps=len(df_test))
    sarima_metrics = calculate_metrics(df_test['price'].values, sarima_forecast)
    print("\nSARIMA Test Metrics:")
    for k, v in sarima_metrics.items():
        print(f"{k}: {v:.4f}")

    # Align LSTM output
    test_index = df_test.index[time_step+1:]
    lstm_aligned = pd.Series(lstm_test_pred.flatten(), index=test_index)

    # Plotting
    plt.figure(figsize=(16, 8))
    plt.plot(df['price'], label='Actual')
    plt.plot(lstm_aligned, label='LSTM')
    plt.plot(df_test.index, arima_forecast, label='ARIMA')
    plt.plot(df_test.index, sarima_forecast, label='SARIMA')
    plt.title("Stock Price Prediction Comparison")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Summary
    print("\nModel Comparison (Test Set):")
    summary = pd.DataFrame({
        'LSTM': {'RMSE': lstm_metrics['RMSE'], 'MAE': lstm_metrics['MAE'], 'MAPE': lstm_metrics['MAPE']},
        'ARIMA': {'RMSE': arima_metrics['RMSE'], 'MAE': arima_metrics['MAE'], 'MAPE': arima_metrics['MAPE']},
        'SARIMA': {'RMSE': sarima_metrics['RMSE'], 'MAE': sarima_metrics['MAE'], 'MAPE': sarima_metrics['MAPE']}
    })
    print(summary)

    best_model = summary.idxmin(axis=1)['RMSE']
    print(f"\n✅ Best model based on RMSE: {best_model}")

# Fix for main function
if __name__ == "__main__":
    main()
```
